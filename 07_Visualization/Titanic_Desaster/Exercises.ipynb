### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv'
titanic = pd.read_csv(url)
titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
titanic.set_index('PassengerId').head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

proportions = [males,females]
plt.pie(
    proportions,
    labels=['Males','Females'],
    shadow=False,
    colors=['blue','red'],
    explode= (0.15,0),
    startangle=90,
    autopct='%.2f%%'
)

plt.axis('equal')
plt.title('Sex Proportions')
plt.tight_layout()
plt.show()
```

```python
titanic.head()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
sns.lmplot(titanic,x='Age',y='Fare',hue='Sex',fit_reg= False)

plt.title('Fare Payed By Age')
plt.ylim(-5,)
plt.xlim(-5,85)
plt.legend()
plt.show()
```

### Step 7. How many people survived?

```python
titanic.Survived.sum()
```

```python
titanic.head()
```

### Step 8. Create a histogram with the Fare payed

```python
to_plt = titanic.sort_values(by='Fare',ascending=False)
bins_val = np.arange(0,600,10)

plt.hist(data=titanic,x='Fare',bins=bins_val,)
plt.ylabel('Frequency')
plt.xlabel('Fare Paid')
plt.title('Fare Paid Histogram')
plt.show()
```

### BONUS: Create your own question and answer it.

```python
#Survivor numbers by cabin 
snc = titanic
snc.dropna()
snc['Cabin_Letter'] = snc['Cabin'].astype(str).str[0]
cabin_survival = snc.groupby('Cabin_Letter')['Survived'].mean().reset_index()
cabin_survival = cabin_survival.sort_values(by="Survived", ascending=False)
sns.barplot(x='Cabin_Letter', y='Survived', data=cabin_survival, palette="coolwarm",hue='Cabin_Letter')
```

```python
cabin_survivalx = snc.groupby('Cabin_Letter')['Survived']
cabin_survival_count  = cabin_survivalx.value_counts().unstack().reset_index()

ind = np.arange(len(cabin_survival_count['Cabin_Letter']))
width = 0.25
x_axis = cabin_survival_count.Cabin_Letter

plt.bar(ind,cabin_survival_count[0],label = 'Not Survived',color = 'red',width=width)
plt.bar(ind + width ,cabin_survival_count[1],label = 'Survived',color = 'blue',width=width)
plt.xticks(ind,x_axis)
plt.title('Survivor and Non Survivor Numbers By Cabin')
plt.legend()
plt.show()
```
